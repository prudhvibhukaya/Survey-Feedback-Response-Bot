```python
import dash
from dash import dcc, html
import plotly.express as px
import pandas as pd


df = pd.read_csv('data/reviews_with_sentiment.csv')


df['sentiment_label'] = df['sentiment_score'].apply(
    lambda x: 'Positive' if x > 0.2 else 'Negative' if x < -0.2 else 'Neutral'
)


app = dash.Dash(__name__)
app.title = "osapiens Sentiment Dashboard"

app.layout = html.Div([
    html.H1("📊 osapiens Sentiment Dashboard", style={'textAlign': 'center'}),

    dcc.Tabs([
        dcc.Tab(label='Sentiment Over Time', children=[
            dcc.Graph(
                figure=px.line(df, x='review_date', y='sentiment_score',
                               title='Sentiment Score Over Time',
                               markers=True)
            )
        ]),
        dcc.Tab(label='Negative Themes', children=[
            dcc.Graph(
                figure=px.histogram(df[df['sentiment_label'] == 'Negative'],
                                    x='module',
                                    title='Negative Feedback by Module',
                                    color_discrete_sequence=['crimson'])
            )
        ]),
        dcc.Tab(label='CSAT vs Feature Usage', children=[
            dcc.Graph(
                figure=px.scatter(df, x='feature_usage_score', y='csat',
                                  color='sentiment_label',
                                  size=abs(df['sentiment_score']), # Use absolute value for size
                                  title='CSAT vs. Feature Usage',
                                  hover_data=['module', 'review_text'])
            )
        ])
    ])
])


if __name__ == '__main__':
    app.run(debug=True)
```
